```python
import pandas as pd
import numpy as np

import scrape_functions 

import time

from tqdm import tqdm
```

```python
nba_abbreviations = [
    'ATL', 'BOS', 'BRK', 'CHI', 'CHO', 'CLE', 'DET', 'IND', 'MIA', 'MIL', 
    'NYK', 'ORL', 'PHI', 'TOR', 'WAS', 'DAL', 'DEN', 'GSW', 'HOU', 'LAC', 
    'LAL', 'MEM', 'MIN', 'NOP', 'OKC', 'PHO', 'POR', 'SAC', 'SAS', 'UTA'
]
```

```python
teams = []

for team in tqdm(nba_abbreviations):
    t = scrape_functions.get_player_ids(team=team, year=2026)
    teams.append(t)
    
    time.sleep(np.random.randint(15,30)) # sleep to prevent lockout 
```

```python
players = {}
for team in teams:
    for player, id in team.items():
        players[player] = id
```
