```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
cursor = connection.cursor()

# Create a table
cursor.execute(""" 
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY, 
        name TEXT, 
        age INTEGER
    )
""")

# Insert data
cursor.execute("""
    INSERT INTO users (name, age) VALUES ('Alice', 30)
""")


connection.commit()
connection.close()
```

```python
with sqlite3.connect("example.db") as conn:
    cur = conn.cursor()

    data = pd.read_sql("""
        SELECT * 
        FROM users
    """, conn)

data
```
